## PRIMERA TÉCNICA: HISTOGRAMAS

**Nombre:** Histogramas

**Origen:** Los histogramas tienen sus raíces en la estadística y la visualización de datos. El término "histograma" fue acuñado en 1891 por el matemático estadístico inglés Karl Pearson

**Descripción/Funcionamiento:** Un histograma es un gráfico de barras que representa la distribución de datos numéricos. Los datos se dividen en intervalos (bins) y se muestra la frecuencia o densidad de datos en cada intervalo mediante barras verticales.

**Ejemplos de aplicación:** Análisis de la distribución de edades en una población, distribución de calificaciones en un examen, análisis de la distribución de ingresos en una muestra demográfica, entre otros.

**Tipo de datos:** Se pueden representar datos cuantitativos (numéricos).

**Limitaciones:**  No hay una medida mínima o máxima estricta, pero la interpretación puede ser limitada si hay muy pocos datos o si los datos están muy dispersos.

**Ejemplo:**

```python
import matplotlib.pyplot as plt

edades = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130, 111, 115, 112, 80, 75, 65, 54, 44, 43, 42, 48]

plt.hist(edades, bins=10, edgecolor='black')
plt.title('Distribución de edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.show()
```

## SEGUNDA TÉCNICA: PEQUEÑAS MULTIPLES

**Nombre:** Pequeñas Múltiples (small multiples)

**Origen:** Introducido por Edward Tufte en su libro "The Visual Display of Quantitative Information". En 1886 Eadward Muybridge Horse in motion, es un ejemplo temprano de pequeñas múltiples.

**Descripción/Funcionamiento:** Consiste en crear múltiples gráficos o visualizaciones pequeñas, cada una representando una parte del conjunto de datos completo. Estas visualizaciones se organizan en una cuadrícula o matriz.

**Ejemplos de aplicación:** Comparación de la evolución de diferentes variables a lo largo del tiempo, comparación de datos demográficos entre diferentes regiones, comparación de desempeño financiero de diferentes empresas, entre otros.

**Tipo de datos:** Se pueden representar datos cuantitativos y cualitativos.

**Limitaciones:** La estructura depende del tipo de gráfico utilizado en cada pequeña visualización. No hay una limitación específica en cuanto a la cantidad de datos, pero el tamaño de cada visualización puede hacer que sea difícil interpretar si hay demasiados datos.

**Ejemplo:**

```python
import matplotlib.pyplot as plt

temperaturas = {
    'Barcelona': [15, 16, 18, 20, 23, 26, 30, 30, 28, 24, 19, 16],
    'Madrid': [10, 12, 15, 18, 22, 30, 35, 35, 30, 20, 15, 10],
    'Bilbao': [10, 11, 13, 15, 18, 20, 22, 22, 21, 18, 14, 11],
    'Valencia': [16, 17, 19, 21, 24, 28, 31, 31, 29, 25, 20, 17]
}

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(meses, temperaturas['Barcelona'], marker='o')
axs[0, 0].set_title('Barcelona')

axs[0, 1].plot(meses, temperaturas['Madrid'], marker='o')
axs[0, 1].set_title('Madrid')

axs[1, 0].plot(meses, temperaturas['Bilbao'], marker='o')
axs[1, 0].set_title('Bilbao')

axs[1, 1].plot(meses, temperaturas['Valencia'], marker='o')
axs[1, 1].set_title('Valencia')

for ax in axs.flat:
    ax.set(xlabel='Mes', ylabel='Temperatura (°C)')

plt.tight_layout()
plt.show()
```

## TERCERA TÉCNICA: Árboles Hiperbólicos

**Nombre:** Árboles Hiperbólicos (Hyperbolic Trees)

**Origen:** Desarrollado para la visualización de estructuras jerárquicas complejas.

**Descripción/Funcionamiento:** Representa relaciones jerárquicas entre elementos en forma de un árbol, utilizando una proyección hiperbólica para mostrar un espacio 2D o 3D que representa la jerarquía de datos.

**Ejemplos de aplicación:** Visualización de estructuras de archivos en un sistema operativo, representación de la jerarquía de categorías en un sitio web, visualización de la estructura de una organización empresarial, entre otros.

**Tipo de datos:** Se utilizan principalmente para representar estructuras jerárquicas cualitativas.

**Limitaciones:** La estructura de los datos debe seguir una jerarquía clara. La limitación principal es la complejidad de la jerarquía, ya que puede ser difícil de interpretar si la estructura es demasiado grande o compleja.

**Ejemplo:**

```python
import matplotlib.pyplot as plt
import networkx as nx

# Definimos la estructura de directorios
estructura_directorios = {
    "nombre": "root",
    "hijos": [
        {
            "nombre": "home",
            "hijos": [
                {
                    "nombre": "user1",
                    "hijos": [
                        {"nombre": "documents"},
                        {"nombre": "downloads"},
                        {"nombre": "pictures"}
                    ]
                },
                {
                    "nombre": "user2",
                    "hijos": [
                        {"nombre": "documents"},
                        {"nombre": "downloads"},
                        {"nombre": "music"}
                    ]
                }
            ]
        },
        {
            "nombre": "etc",
            "hijos": [
                {"nombre": "nginx"},
                {"nombre": "ssh"}
            ]
        },
        {
            "nombre": "var",
            "hijos": [
                {"nombre": "log"},
                {"nombre": "lib"},
                {"nombre": "cache"}
            ]
        }
    ]
}

# Creamos una función recursiva para añadir nodos y aristas al gráfico
def add_nodes(graph, node_data, parent=None):
    node_name = node_data["nombre"]
    graph.add_node(node_name)
    if parent is not None:
        graph.add_edge(parent, node_name)
    for child in node_data.get("hijos", []):
        add_nodes(graph, child, parent=node_name)

# Creamos un gráfico dirigido y añadimos los nodos y aristas
G = nx.DiGraph()
add_nodes(G, estructura_directorios)

# Dibujamos el gráfico
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray')

plt.show()
```
